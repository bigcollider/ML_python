```python
import numpy as np
import pandas as pd
from scipy.stats.stats import pearsonr

df = pd.read_csv('f:/programming/data/titanic.csv', index_col='PassengerId')
#print (df.head(5))
#df.info()

#1.sex
print('1.SEX: ', df.groupby('Sex').size())

#2.percent of survived passengers
count_all_pas = df['Survived'].count()
count_surv_pas = df['Survived'][df['Survived']==1].count()
pt_surv_pas = round(count_surv_pas/count_all_pas,2)
print('2. all pass: {0}; surv pass: {1}; percent: {2}'.format(count_all_pas, count_surv_pas,pt_surv_pas))

#3. percent of 1st class passengers
count_firstC_pas = df['Pclass'][df['Pclass']==1].count()
pt_firstC_pas = round(count_firstC_pas/count_all_pas,2)
print('3. all pass: {0}; first pass: {1}; percent: {2}'.format(count_all_pas, count_firstC_pas,pt_firstC_pas))

#4.mean and median of age
print('4.')
print(df['Age'].agg([np.mean,np.median]))

#5. correlation
print('5. ', pearsonr(df['SibSp'],df['Parch']))
      


```
